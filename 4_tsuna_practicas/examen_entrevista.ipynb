```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\tsuda\\dataelite\\dataelite2025\\Tsuna_practicas\\tabla.csv")
df.head(10)
```

```python
df.dtypes
```

```python
df.duplicated().sum()  # Duplicados
```

```python
df.isnull().sum()  # nulos
```

```python
df.columns = df.columns.str.lower().str.replace("_"," ").str.strip()  # envia los nombres de las columnas a minusculas, reemplaza _
df.columns = df.columns.str.strip().str.replace("numer0 telefonico", "numero telefonico")  # cambiar nombre de la columna
```

```python
df["fecha"]
```

```python
df["fecha"] = pd.to_datetime(df["fecha"], errors= "coerce")
```

```python
# limpiar genero
df["genero"] = df["genero"].str.lower().str.strip().replace({"mujer":"femenino", "f":"femenino", r"(^m$)":"masculino"}, regex = True)
```

```python
df
```

```python
#limpiar estado civil
df["estado civil"] = df["estado civil"].str.lower().str.replace("_"," ").str.strip()
df
```

```python
#limpiar nombre
df["nombre"] = df["nombre"].str.lower().str.replace("."," ").str.replace("_", " ").str.strip()
```

```python
df["numero telefonico"]=df["numero telefonico"].astype(str).apply(lambda x: x if len(str(x)) == 10 else None)
```

```python
df.dtypes
```

```python
df
```

```python
df = df.drop("id", axis=1)
```

```python
df = df.drop_duplicates()
```

```python
df["id"] = df.index +1
```
